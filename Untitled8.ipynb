```python
import pandas as pd

df = pd.read_csv("/content/data.xlsx - zomato.csv")
```

```python

print(df.head())


print(df.info())

print(df.describe())

print(df.columns)
```

```python
print(df.isnull().sum())

df.dropna(subset=['cuisines', 'Rating', 'votes', 'location'], inplace=True)

df['approx_cost(for two people)'].fillna(df['approx_cost(for two people)'].median(), inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['approx_cost(for two people)'] = df['approx_cost(for two people)'].astype(str).str.replace(',', '')
df['approx_cost(for two people)'] = df['approx_cost(for two people)'].astype(float)


df['Rating'] = df['Rating'].astype(str).str.replace('/5', '').str.strip()
df = df[df['Rating'] != 'NEW']
df = df[df['Rating'] != '-']
df['Rating'] = df['Rating'].astype(float)

df['location'] = df['location'].str.strip().str.title()
df['cuisines'] = df['cuisines'].str.strip().str.title()
```

```python
df['Rating'] = df['Rating'].astype(float)
df['votes'] = df['votes'].astype(int)
df['approx_cost(for two people)'] = df['approx_cost(for two people)'].astype(float)
```

```python
df.to_csv("zomato_cleaned.csv", index=False)
```

```python
grouped = df.groupby('cuisines').agg({
    'Rating': 'mean',
    'votes': 'mean',
    'approx_cost(for two people)': 'mean'
}).sort_values(by='Rating', ascending=False)

print(grouped.head(10))
```

```python
top_by_rating = grouped.sort_values(by='Rating', ascending=False).head(10)
top_by_votes = grouped.sort_values(by='votes', ascending=False).head(10)
```

```python
top_restaurants = df[df['Rating'] >= 4.5][['name', 'cuisines', 'Rating', 'votes', 'location']]
print(top_restaurants.sort_values(by='Rating', ascending=False).head(10))
```

```python
df[df['location'] == 'Bangalore'].groupby('cuisines').mean()
```

```python
grouped = df.groupby('cuisines').agg({
    'Rating': 'mean',
    'votes': 'mean',
    'approx_cost(for two people)': 'mean'
})

top_by_rating = grouped.sort_values(by='Rating', ascending=False).head(10)

top_by_votes = grouped.sort_values(by='votes', ascending=False).head(10)
```

```python
import matplotlib.pyplot as plt

top_by_rating['Rating'].plot(kind='bar', title='Top Cuisines by Rating')
plt.ylabel('Average Rating')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

top_by_votes['votes'].plot(kind='bar', title='Top Cuisines by Votes', color='green')
plt.ylabel('Average Votes')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
